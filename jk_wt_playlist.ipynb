```python
browser=webdriver.Chrome()
```

```python
browser.get(link)
```

```python
soup=BeautifulSoup(browser.page_source,"html.parser")
```

```python
soup.text.replace("\n\n",'')
```

```python
soup.find_all("ytd-playlist-panel-renderer",class_="style-scope ytd-watch-flexy")[0]
```

```python
videos_data=[]
number=1
for ind_video in tqdm(soup.find_all("ytd-playlist-panel-renderer",class_="style-scope ytd-watch-flexy")[0].find_all("ytd-playlist-panel-video-renderer",class_="style-scope ytd-playlist-panel-renderer")):
    links="https://www.youtube.com/"+ind_video.find("a").get("href").split("&")[0]
    title=ind_video.find("h4",class_="style-scope ytd-playlist-panel-video-renderer").find("span",class_="style-scope ytd-playlist-panel-video-renderer").text.strip()
    duration=ind_video.find("span",class_="style-scope ytd-thumbnail-overlay-time-status-renderer").text.replace("\n","").strip()
    
    videos_data.append([number,title,duration,links])
    number+=1
```

```python
df=pd.DataFrame(videos_data,columns=["NUMBER","TITLE","DURATION","LINK"],index=None)
```

```python
df
```

```python
df.to_csv("jk_wt_playlist.csv",index=False)
```

```python
timings=[]
for time in df["DURATION"]:
    t=time.split(":")
    m=":".join(t)
    timings.append(m)
```

```python
df["DURATION"]=timings
```
